```python
import pandas as pd
import MySQLdb
import configparser

DB_CONF = '/home/slurm/slurm_sim_ws/sim/micro/baseline/etc/slurmdbd.conf'

def slurm_db():
    # обработка конфига
    slurmdbdconf_str = "[fake_header]\n" + open(DB_CONF, 'r').read()
    slurmdbdconf = configparser.RawConfigParser()
    slurmdbdconf.read_string(slurmdbdconf_str)
    db_login = slurmdbdconf.get("fake_header", "StorageUser")
    db_password = slurmdbdconf.get("fake_header", "StoragePass")
    db_host = slurmdbdconf.get("fake_header", "DbdHost")
    db_port = slurmdbdconf.get("fake_header", "DbdPort")
    db_storage = slurmdbdconf.get("fake_header", "StorageLoc")

    db = MySQLdb.connect(
        host=db_host,
        user=db_login,
        passwd=db_password,
        port=int(db_port),
        db=db_storage
    )

    cursor = db.cursor(MySQLdb.cursors.DictCursor)
    query = f"SHOW TABLES"
    cursor.execute(query)
    data = cursor.fetchall()

    # Создаем dataframe из данных
    df = pd.DataFrame(data)
    for x in df.iloc[:,0]:
        query = f"select * from {x}"
        cursor.execute(query)
        data = cursor.fetchall()
        df = pd.DataFrame(data)
        print(x, df, end='\n\n\n')
        
        
    # сохраняем dataframe в csv
    df.to_csv('cluster_job_table.csv', index=False)

    db.close()

    # print(df)
    return df
```

```python
df = slurm_db()
```
